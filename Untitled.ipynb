```python
import pandas as pd
import panel as pn
import numpy as np
import matplotlib.pyplot as plt
import bokeh.plotting as bk
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
pn.extension(template = 'fast')
```

```python
df = pd.DataFrame(
    {
        'A': [1,2,3,4,5],
        'B': [6,7,8,9,10]
    }
)

choose_column = pn.widgets.Select(name='Column', value='A', options=list(df.columns))

def display(column):
    return df[column]

pn.Column(choose_column, pn.panel(pn.bind(display, choose_column)))
```

```python
start_slide = pn.widgets.IntSlider(name='Start', start=0, end=100, value=0)
end_slide = pn.widgets.IntSlider(name='End', start=0, end=100, value=100)

def numbers(start, end):
    df = pd.Series([])
    for i in range(start, end):
        df.loc[len(df)]=i
    return df

pn.Column(
    start_slide,
    end_slide,
    pn.panel(pn.bind(numbers, start_slide, end_slide))
)
```

```python
x_slider_begin = pn.widgets.FloatSlider(name='x_start', value=0, start=-9.0, end=9.0)
x_slider_end = pn.widgets.FloatSlider(name='x_end', value=0, start=-9.0, end=9.0)

def func(x):
    return x**2

def plot_curve(x_begin, x_end):
    x = np.linspace(x_begin, x_end, 100)
    y = func(x)
    source = ColumnDataSource(data={'x': x, 'y': y})
    
    p = figure(title='Plot of y = x^2', x_axis_label='x', y_axis_label='y')
    p.circle('x', 'y', source=source, line_width=2)
    
    return p

pn.Column(
    x_slider_begin,
    x_slider_end,
    pn.Row(pn.panel(pn.bind(plot_curve, x_slider_begin, x_slider_end)), pn.panel(pn.bind(plot_curve, x_slider_begin, x_slider_end)))
).servable()
```
